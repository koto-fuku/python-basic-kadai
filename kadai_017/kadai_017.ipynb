```python
# Humanクラスの定義
class Human:

    # コンストラクタ
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 成人判定メソッド
    def check_adult(self):
        if self.age >= 20:
            print(self.name + "さんは大人です")
        else:
            print(self.name + "さんは大人ではありません")


# Humanインスタンスを作成する関数
def create_humans():

    # ディクショナリのリスト
    data_list = [
        {"name": "KOTOMI", "age": 30},
        {"name": "TARO", "age": 18},
        {"name": "HANAKO", "age": 22}
    ]

    human_list = []

    # 繰り返し処理
    for data in data_list:
        human = Human(data["name"], data["age"])
        human_list.append(human)

    return human_list


# メイン処理
humans = create_humans()

# リストの要素数分メソッド実行
for person in humans:
    person.check_adult()
```
